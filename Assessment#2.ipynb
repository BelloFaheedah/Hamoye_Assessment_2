<h2>Multiple Linear Regression</h2>
<p>Using the Appliances Energy Data</p>

```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing the dataset
mydata=pd.read_csv('energydata_complete.csv')
mydata
```

```python
#dropping the columns 'date' and 'lights' from the dataset
my_new_data=mydata.drop(columns=['lights', 'date'])
my_new_data
```

```python
#normalising the dataset
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_data = pd.DataFrame(scaler.fit_transform(my_new_data), columns=my_new_data.columns)
target = normalised_data['Appliances']
predictors=normalised_data.iloc[:,1:27]
```

```python
#splitting the dataset into training and testing in the ratio 70:30 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
```

```python
#checking for multicollinearity among the predictor variables
predictors.corr()
```

```python
predictors.corr().style.background_gradient(cmap='coolwarm')
```

<p>The table, especially the figure above shows it, that most variables have a high correlation among themselves, (The darker the red, the higher the correlation), and this indicates multicollinearity</p>

<h4>Linear Regression</h4>
<p>Fitting the linear model on the normalised dataset</p>

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
#fitting the model to the training dataset
linear_model.fit(x_train, y_train)
```

```python
#obtain predictions, evaluating the model on the test set
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
rmse
```

```python
from sklearn.metrics import r2_score
r2_score_linear = r2_score(y_test, predicted_values)
r2_score_linear
```

<p>The coefficient of determination explains that 14.89% of the variation in the Appliances is predicted by the predictor variables</p>

<h4>Ridge regression</h4>
<p>In order to take care of the anomaly (i.e. the multicollinearity among the predictor variables)</p>

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(x_train, y_train)
```

```python
ridge_pred=ridge_reg.predict(x_test)
```

```python
from sklearn.metrics import  mean_squared_error
rmse_ridge = np.sqrt(mean_squared_error(y_test, ridge_pred))
rmse_ridge
```

```python
from sklearn.metrics import r2_score
r2_score_ridge = r2_score(y_test, ridge_pred)
r2_score_ridge
```

<h4>Lasso regression </h4>
<p>Correcting the anomaly of multicollinearity</p>

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_pred=lasso_reg.predict(x_test)
```

```python
from sklearn.metrics import  mean_squared_error
rmse_lasso = np.sqrt(mean_squared_error(y_test, lasso_pred))
rmse_lasso
```

```python
from sklearn.metrics import r2_score
r2_score_lasso = r2_score(y_test, lasso_pred)
r2_score_lasso
```

<h4>Elastic Net Regression</h4>
<p>Balancing between lasso and ridge regression</p>

```python
from sklearn.linear_model import ElasticNet
from sklearn.datasets import make_regression
predictors, target = make_regression(n_features=26, random_state=42)
regr = ElasticNet(random_state=42)
regr.fit(x_train, y_train)
```

```python
from sklearn.metrics import  mean_squared_error
elastic_pred=regr.predict(x_test)
rmse_elastic = np.sqrt(mean_squared_error(y_test, elastic_pred))
rmse_elastic
```

```python
from sklearn.metrics import r2_score
r2_score_elastic = r2_score(y_test, elastic_pred)
r2_score_elastic
```

```python
def get_weights_df(linear_model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(linear_model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
elastic_weights_df =get_weights_df(regr, x_train,'Elastic_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
final_weights= pd.merge(final_weights,elastic_weights_df, on='Features')
```

```python
final_weights
```

```python
#comparing the models
compare=pd.DataFrame(['Linear model', 'Ridge regression', 'Lasso regression', 'Elastic Net Regression'], columns=['Models'])
compare['R_squared']=[r2_score_linear,r2_score_ridge,r2_score_lasso,r2_score_elastic]
compare['RMSE'] = [rmse,rmse_ridge,rmse_lasso,rmse_elastic]
compare
```

<p>it is reaized that <b>Ridge Regression</b>performed better amidst the other three. Even though <b>Linear regression</b> has better performance values, it is however not good to use since multicollinearity is confirmed in the dataset</p>
